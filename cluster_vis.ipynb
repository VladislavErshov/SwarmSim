```python
from sklearn.datasets import make_blobs, make_moons, make_circles
from matplotlib import pyplot
from pandas import DataFrame
```

```python
titles = ['blobs', 'moons', 'circles']
```

```python
X, y = make_blobs(n_samples=1000, centers=3, cluster_std=0.9, n_features=2)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue', 2:'green'}
grouped = []
grouped.append(df.groupby('label'))
```

```python
X, y = make_moons(n_samples=1000, noise=0.05)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
grouped.append(df.groupby('label'))
```

```python
X, y = make_circles(n_samples=1000, noise=0.02)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
grouped.append(df.groupby('label'))
```

```python
fig, ax = pyplot.subplots(1, len(grouped), figsize=(6, 2), dpi=200)
for idx, grp in enumerate(grouped):
    for key, group in grp:
        group.plot(ax=ax[idx], kind='scatter', x='x', y='y', label=key, color=colors[key], s=1)
        if idx > 0:
            ax[idx].set_ylabel('')
        ax[idx].set_title(titles[idx])
#fig.tight_layout()
fig.tight_layout(pad=0.1)
pyplot.show()
```
